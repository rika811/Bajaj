```python

import pandas as pd
import json


with open('/content/DataEngineeringQ2.json','r') as file:
    data = json.load(file)


df = pd.json_normalize(data, sep='_')

patient_details_cols = [col for col in df.columns if 'patientDetails' in col]
consultation_data_cols = [col for col in df.columns if 'consultationData' in col]

df_patient_details = df[patient_details_cols]
df_consultation_data = df[consultation_data_cols]


print(df_patient_details.head())
print(df_consultation_data.head())

df['medicine_count'] = df['consultationData_medicines'].apply(lambda x: len(x))


df['total_medicine_duration'] = df['consultationData_medicines'].apply(lambda meds: sum(int(med['duration']) for med in meds if med['duration']))

print(df[['appointmentId', 'medicine_count', 'total_medicine_duration']].head())
```

```python
import pandas as pd
import json


with open('/content/DataEngineeringQ2.json', 'r') as file:
    data = json.load(file)

df = pd.json_normalize(data, sep='_')

def is_valid_indian_number(phone_number):
    if isinstance(phone_number, str):

        if phone_number.startswith('+91'):
            phone_number = phone_number[3:]
        elif phone_number.startswith('91'):
            phone_number = phone_number[2:]


        if phone_number.isdigit() and len(phone_number) == 10 and phone_number[0] in '6789':
            return True
    return False

df['isValidMobile'] = df['patientDetails_phoneNumber'].apply(is_valid_indian_number)

valid_phone_numbers_count = df['isValidMobile'].sum()

valid_phone_numbers_count
```

```python
import pandas as pd
import json


with open('/content/DataEngineeringQ2.json', 'r') as file:
    data = json.load(file)


df = pd.json_normalize(data, sep='_')


def count_medicines(medicines):
    if isinstance(medicines, list):
        return len(medicines)
    return 0


df['medicine_count'] = df['consultationData_medicines'].apply(count_medicines)


df['patientDetails_age'] = pd.to_numeric(df['patientDetails_age'], errors='coerce')


df = df.dropna(subset=['medicine_count', 'patientDetails_age'])


pearson_corr = df['medicine_count'].corr(df['patientDetails_age'])

pearson_corr
```
